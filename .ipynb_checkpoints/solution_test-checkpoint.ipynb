```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Part 1: Data Extraction, Cleaning, and Transformation 
# 1. Data Extraction
# Load CSV Data using pandas
patient_demographics = pd.read_csv('patient_demographics.csv')
patient_visits = pd.read_csv('patient_visits.csv')
patient_lab_results = pd.read_csv('patient_lab_results.csv')
patient_medications = pd.read_csv('patient_medications.csv')
physician_assignments = pd.read_csv('physician_assignments.csv')
```

```python
# 2. Data Cleaning
# Handle missing values
patient_demographics.isna().sum()
```

```python
#distribution légèrement asym vers la droite, la moyenne pourrait 
#être biaisée, on utilisera la médiane de la colonne pour remplacer les nan
```

```python
patient_demographics["age"].fillna(patient_demographics["age"].median(),inplace=True)
patient_demographics['gender'].fillna("Unknown", inplace=True) #remplir avec "Unknown" pr eviter tt prob lié à la fusion
patient_demographics['other_fields'].fillna('No information', inplace=True)   #ne pas biaiser les donnees
```

```python
# No duplicated records detected 

# Normalize date formats 
patient_visits['visit_date'] = pd.to_datetime(patient_visits['visit_date'])
physician_assignments['assignment_date'] = pd.to_datetime(physician_assignments['assignment_date'])
patient_lab_results['test_date']=pd.to_datetime(patient_lab_results['test_date'])
patient_medications['start_date']=pd.to_datetime(patient_medications['start_date'])
patient_medications['end_date']=pd.to_datetime(patient_medications['end_date'])

#No need to normalize the texts ,already normalized
```

```python
# Data Transformation & Merging:
# Merge datasets on common keys .
df = patient_demographics.merge(patient_visits, on="patient_id", how="left")
df = df.merge(patient_medications, on=["patient_id", "visit_id"], how="left")
df = df.merge(patient_lab_results, on=["patient_id", "visit_id"], how="left")
df = df.merge(physician_assignments, on=["patient_id", "visit_id"], how="left")

#evitons toute confusion en renommant les colonnes medicaments et other fields
df.rename(columns={
    "medication_x": "diagnosed_medication",  # Médicament diagnostiqué lors de la visite
    "medication_y": "prescribed_medication",  # Médicament réellement prescrit
    "other_fields_x": "patient_notes",  # Infos du patient (ex: Non-smoker, Diabetic)
    "other_fields_y": "visit_notes",  # Infos de la visite (ex: Follow-up, Initial assessment)
    "notes_x": "medication_notes",
    "notes_y": "test_notes"
}, inplace=True)

# Création d'une colonne "Nombre de visites"
df["visit_count"] = df.groupby("patient_id")["visit_id"].transform("count")

# Classification des âges en groupes
bins = [0, 18, 35, 50, 65, 100]
labels = ["Enfant", "Jeune adulte", "Adulte", "Senior", "Âgé"]
df["age_group"] = pd.cut(df["age"], bins=bins, labels=labels, right=False)

# Calcul de la durée des traitements
df["medication_duration"] = (pd.to_datetime(df["end_date"]) - pd.to_datetime(df["start_date"])).dt.days
```

```python
columns_with_nan = df.columns[df.isna().any()]

# Afficher les colonnes avec NaN
df[columns_with_nan]
```

```python
# Traitement des NaN pour les médicaments
df["medication_id"].fillna("Unknown", inplace=True)
df["prescribed_medication"].fillna("No medication prescribed", inplace=True)
df["dosage"].fillna("No dosage information", inplace=True)
df["start_date"].fillna("1970-01-01", inplace=True)
df["end_date"].fillna("Not available", inplace=True)
df["medication_notes"].fillna("No notes available", inplace=True)
df["medication_duration"].fillna("No information", inplace=True)

# Traitement des NaN pour les tests de laboratoire
df["lab_test_id"].fillna("Unknown", inplace=True)
df["test_date"].fillna("1970-01-01", inplace=True)
df["test_name"].fillna("No test performed", inplace=True)
df["result_value"].fillna("No result", inplace=True)
df["result_unit"].fillna("No unit", inplace=True)
df["reference_range"].fillna("No reference_range", inplace=True)
df["test_notes"].fillna("No notes available", inplace=True)

# Vérification des NaN après le traitement
print(df.isna().sum())
```

```python
# Vérifier si toutes les lignes ont un diagnosed_medication mais pas de medication_id
df_missing_med_id = df[(df["diagnosed_medication"].notna()) & (df["medication_id"].isna())]
print("Cas où un médicament est diagnostiqué mais non prescrit :")
print(df_missing_med_id)

# Vérifier s'il y a des valeurs NaN dans medication_duration
print("Lignes avec medication_duration manquant :")
print(df[df["medication_duration"].isna()])

# Vérifier si des patients ont plusieurs médecins pour une même visite
df_duplicate_physicians = df.duplicated(subset=["patient_id", "visit_id", "physician_id"], keep=False)
print("Patients avec plusieurs médecins pour une visite :")
print(df[df_duplicate_physicians])
```
